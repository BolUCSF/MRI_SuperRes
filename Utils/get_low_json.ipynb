```python
new_list = []
for image_path in glob(f'/working/Data/Images/Cerebro_data/nifti_fl_bet/*_bet.nii.gz'):
    nib_img = nib.load(image_path)
    data = nib_img.get_fdata()
    
    if data.shape[2] < 100:
        print(data.shape)
        print(image_path)
        print(nib_img.header['pixdim'][1:4])
        new_list.append({"fl":image_path})
    if len(new_list) == 100:
        break
```

```python
with open('/working/Project/Sup_Res/json/short_list.json', 'w') as f:
    json.dump(new_list, f, indent=4)
```
