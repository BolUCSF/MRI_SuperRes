```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fl_bicubic = pickle.load(open('fl_bicubic_3.pkl', 'rb'))
fl_down = pickle.load(open('fl_downsample_3.pkl', 'rb'))
fl_recon = pickle.load(open('fl_recon_3.pkl', 'rb'))
#  3 
data_dict = {
    "Nearest neighbor": fl_down['psnr'],
    "Bicubic": fl_bicubic['psnr'],
    "Superformer": fl_recon['psnr']
}

#  Seaborn 
data = []
categories = []
means = []  # 
for key, values in data_dict.items():
    data.extend(values)
    categories.extend([key] * len(values))
    means.append(np.mean(values))  # 

#  Violin 
plt.figure(figsize=(8, 5))
ax = sns.violinplot(x=categories, y=data, inner="quartile", palette="Set2")

# 
for i, mean_value in enumerate(means):
    plt.scatter(i, mean_value, color="red", marker="o", s=50, label="Mean" if i == 0 else "")
    plt.text(i, mean_value + 0.5, f"{mean_value:.2f}", ha="center", va="bottom", fontsize=10, color="red")

# 
# plt.xlabel("Category")
plt.ylabel("PSNR")
plt.title("Downsample rate 3 on FLAIR")
plt.legend()  # 
plt.grid(axis="y", linestyle="--", alpha=0.7)

# 
plt.show()
```
